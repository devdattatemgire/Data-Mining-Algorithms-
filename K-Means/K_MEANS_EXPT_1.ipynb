#**STATIC** **1D**

```python
import statistics
import numpy as np
min = 999
k = int(input("Enter the value for k"))
D = [2,3,4,10,11,12,20,25,30]
dist = np.empty(shape=(int(len(D)),int(k)))
dist.fill(0)
# print(dist)
oldcent = []
newcent = []


clusters = [[] for _ in range(k)]
print(clusters)
for i in range(int(k)):
  oldcent.append(int(input("Enter the Centroid")))
print("oldcent: ",oldcent)
print("clusters: ", clusters)
while(newcent != oldcent):
    clusters = [[] for _ in range(k)]
    if all(element != 0 for element in newcent):
      oldcent = newcent.copy()
    for i in range(int(len(D))):
      for j in range(k):
       dist[i,j] = abs(D[i]-oldcent[j])
    print("Distance: ",dist)
    for i in range(int(len(D))):
      for j in range(k):
       if dist[i,j] < min:
        min = dist[i,j]
        index = j
        clusters[index].append(min)
    print("Clusters Array: ", clusters)
    for cluster in clusters:
      mean_value = statistics.mean(cluster)
      newcent.append(mean_value)
    print("newcent: ", newcent)
```

#**DYNAMIC** **2D**


```python
import statistics
import numpy as np
import math

min_dist = 999
k = int(input("Enter the value for k"))
D =  [(2, 3), (3, 5), (5), 8, (10, 12), (12, 15), (14, 18), (20, 22), (25, 30), (30, 35)]
dist = np.empty(shape=(int(len(D)), int(k)))
dist.fill(0)

oldcent = []
newcent = []

clusters = [[] for _ in range(k)]
print(clusters)

for i in range(int(k)):
    x, y = input("Enter two values: ").split()
    oldcent.append([int(x),int(y)])


print("oldcent: ", oldcent)
print("clusters: ", clusters)
count = 0
while newcent != oldcent:
    clusters = [[] for _ in range(k)]
    min_dist = 999
    if count != 0:
     oldcent = newcent.copy()
     print("oldcent: ", oldcent)
    newcent = []

    for i in range(int(len(D))):
        for j in range(k):
            # dist[i, j] = abs(D[i] - oldcent[j])
            dist[i, j] = math.dist(D[i],oldcent[j])

    print("Distance: ", dist)

    for i in range(int(len(D))):
        min_dist = 999
        for j in range(k):
            if dist[i, j] < min_dist:
                min_dist = dist[i, j]
                index = j
        clusters[index].append(D[i])

    print("Clusters Array: ", clusters)

    for cluster in clusters:
      print("cluster: ", cluster)
      x_sum = sum(point[0] for point in cluster)
      y_sum = sum(point[1] for point in cluster)
      mean_value_x = x_sum / len(cluster)
      mean_value_y = y_sum / len(cluster)
      newcent.append((mean_value_x, mean_value_y))
    print("newcent: ", newcent)
    count = count + 1
```
