```python
import pandas as pd
from statistics import mean
import math
df = pd.read_csv('/content/drive/MyDrive/salary data.csv')
```

```python
X = df.drop(df.columns[[-1]], axis = 1)
y = df.iloc[:,-1:]
# print(X)

# min = X.iloc[:,0:].min()
# max = X.iloc[:,0:].max()
# X.iloc[:,0:] = (X.iloc[:,0:] - min) / (max - min)
```

```python
x = X.iloc[:, -1].tolist()
Y = y.iloc[:, -1].tolist()
```

```python
def predict(x_known):

  x_xmean = []
  y_ymean = []

  for i,j in zip(x,Y):
    x_xmean.append(i-mean(x))
    y_ymean.append(j-mean(Y))
  x_y = 0
  x_xmean_sqr = 0
  for i,j in zip(x_xmean, y_ymean):
    x_y = x_y + (i * j)
    x_xmean_sqr = x_xmean_sqr + (i**2)

  b1 = x_y / x_xmean_sqr
  b0 = mean(Y) - b1*mean(x)
  y_unknown = b0 + b1* x_known

  # print(b0,b1)

  return y_unknown

```

```python
x_known = int(input("Enter the Years of Experience: "))
predict(x_known)
```

```python
se = []
y_pred = []
for i,j in zip(x,Y):
  print("y_test: ", j )
  print("y_pred:", predict(i))
  y_pred.append(predict(i))
  print('\n')
  se.append((j-predict(i))**2)
mse = mean(se)

print('Mean Squared Error: ',mse)
print('Root Mean Squared Error: ',math.sqrt(mse))
```

```python
yd_pred = pd.DataFrame(y_pred)
```

```python
import matplotlib.pyplot as plt
# Plot the data points
plt.scatter(X, y, label='Data Points')

# Plot the regression line
plt.plot(X, yd_pred, color='red', linewidth=2, label='Regression Line')

# Add labels and legend
plt.xlabel('Input Feature')
plt.ylabel('Target')
plt.legend()

# Show the plot
plt.show()
```
